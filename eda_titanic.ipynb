## **2.3 Meta Data:**
The dataset contains information about 891 entries with the following columns:

1. **Survived**: Indicates whether the passenger survived (1) or not (0).
2. **Pclass**: Passenger class (1 = 1st, 2 = 2nd, 3 = 3rd).
3. **Sex**: Gender of the passenger (male/female).
4. **Age**: Age of the passenger.
5. **Sibsp**: Number of siblings/spouses aboard the Titanic.
6. **Parch**: Number of parents/children aboard the Titanic.
7. **Fare**: Passenger fare.
8. **Embarked**: Port of embarkation (C = Cherbourg; Q = Queenstown; S = Southampton).
9. **Class**: Class of the passenger (First, Second, Third).
10. **Who**: Description of the passenger (man, woman, child).
11. **Adult_male**: Indicates if the passenger is an adult male (True/False).
12. **Deck**: Deck where the passenger was located.
13. **Embark_town**: Town where the passenger boarded.
14. **Alive**: Indicates whether the passenger survived (yes/no).
15. **Alone**: Indicates if the passenger was alone (True/False).

# 5.3 Let's solution to the missing values

```python
# Handle missing values

# Fill the missing values of age column by the mean
df['age'].fillna(df['age'].mean(), inplace = True)

# Fill the missing values of embarked column by the mode
df['embarked'].fillna(df['embarked'].mode()[0], inplace = True)

# Fill the missing values of embark_town column by the mode
df['embark_town'].fillna(df['embark_town'].mode()[0], inplace = True)
```

# 5.4 Let's Remove the Deck Column

```python
# 1 -drop the columns haveing more than 77% of missing values from dataset
df.drop('deck', axis = 1, inplace = True)
```

# Observation set: 5

Based on our analysis, we have observed that the **`Deck`**, **`Age`**, **`Embarked`**, and **`Embark_town`** columns have missing values. As data analysts, we recommend the following corrections and insights:

- **Deck**:
  - **Missing Instances**: `688`
  - **Recommendation**: We have decided to drop the `Deck` column due to its `77%` missing data, making accurate imputation unreliable. Removing this column will simplify the dataset, enhance manageability, and ensure a more robust and straightforward analysis.

- **Age**:
  - **Missing Instances**: `177`
  - **Recommendation**: Impute the missing age values using the median or mean age. This approach helps to retain the usefulness of the `Age` column in your analysis while maintaining data integrity.

- **Embarked**:
  - **Missing Instances**: `2`
  - **Recommendation**: Impute the missing values in the `Embarked` column using the most frequent port of embarkation (mode). This approach minimizes the impact on your analysis and maintains data completeness.

- **Embark_town**:
  - **Missing Instances**: `2`
  - **Recommendation**: Similarly, impute the missing values in the `Embark_town` column using the most frequent town of embarkation. This helps maintain consistency and completeness with minimal disruption.

By addressing these missing values appropriately, we aim to ensure the quality and reliability of the dataset for further analysis.

# Let's Do Feature Engineering 

```python
# Let's binning of the age column 

bins = [0,1,5,12,18,30,50,80]
labels = ['Infants', 'Toddlers', 'Kids', 'Teens', 'Youngs Adult', 'Adult', 'Senior']

# Let's make a new column of age group
df['age_group'] =pd.cut(df['age'], bins = bins, labels = labels) 
```

```python
# Let's make a new column of family size
df['family_size'] = df['sibsp'] + df['parch'] + 1  # 1 is for the person himself 
```

# Univariate Analysis

### Q 1: What does the distribution of ages among passengers look like?

```python
# Age distribution
plt.figure(figsize=(16, 6))
sns.histplot(df['age'], kde=True)
plt.title('Age Distribution')
plt.show()
```

**Figure-3** This histogram with a KDE shows the `Age distribution` of the passengers.

**Answer**: The plot reveals peaks among young adults and adults, with fewer infants, toddlers, and seniors, providing insights into the age demographics of the passengers.

### Q 2: How are passengers distributed across different classes?

```python
# Class distribution
plt.figure(figsize=(16, 6))
sns.countplot(x='pclass', data=df, palette='husl')
plt.title('Class Distribution')
plt.show()
```

**Figure-4** shows the **`Class`** distribution of the passengers.

**Answer**: The count plot indicates the number of passengers in each **`Class`**, highlighting that most passengers were in the **`Third class`**, followed by the first and second classes.

### Q 3: What is the distribution of family sizes among passengers?

```python
# Family size distribution using a bar plot
plt.figure(figsize=(16, 6))
sns.countplot(x='family_size', data=df, palette='husl')
plt.title('Family Size Distribution')
plt.xlabel('Family Size')
plt.ylabel('Count')
plt.show()
```

**Figure-5** shows the **`Family Size`** distribution of the passengers using a bar plot.

**Answer**: The bar plot illustrates the number of passengers for each family size category, clearly showing that most passengers traveled alone (family size of 1) or with small families (family size of 2 or 3).
The number of passengers decreases as the family size increases, with very few passengers having a family size of 7 or more.

### Q 4: What is the distribution of survival outcomes among passengers?

```python
#Survival Distribution Plot
plt.figure(figsize=(16, 6))
sns.countplot(x='survived', data=df, palette='Set1')
plt.title('Survival Distribution')
plt.xlabel('Survived ( 0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

**Figure-6** The plot displays the **`Distribution of Survival`** of passengers

**Answer**: The customized bar plot distinguishes between passengers who did not survive **`(0 = No)`** and those who survived **`(1 = Yes)`**. It visually emphasizes the higher count of non-survivors compared to survivors, offering a straightforward view of the survival distribution aboard the Titanic.

#  Bivariate Analysis

### Q 5: How does age correlate with survival outcomes? 

```python
# Age Distribution of Survival and Non Survival
plt.figure(figsize=(16, 6))
sns.histplot(data=df, x='age', hue='survived', multiple='stack', bins=30, kde=True)
plt.title('Age Distribution of Survival vs Non Survival')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['0', '1'])
plt.show()
```

**Figure-7** The plot shows the Age Distribution of survival and non survival.

**Answer**: The bar plot illustrates the number of passengers `who survived (1)` and those `who did not survive (0)`. It shows that there were more passengers who did not survive compared to those who survived.

### Q 6 : How does passenger class affect survival rates, considering gender and age group?

```python
#class vs survival and gender 
plt.figure(figsize=(16, 6))
sns.barplot(x='pclass', y='survived', hue='who', data=df, palette='tab10')
plt.title('Class vs Survival by Gender and Age Group')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')

# Get the current legend handles and labels
handles, labels = plt.gca().get_legend_handles_labels()
# Set the new labels
new_labels = ['Female Adults', 'Male Adults', 'Children']
plt.legend(handles=handles, labels=new_labels, title='Gender and Age Group')
plt.show()
```

**Figure-8** The plot shows that the survival rates across passenger classes categorized by gender and age group.

**Answer**: The bar plot demonstrates that female adults had the highest survival rates across all passenger classes, followed by children and male adults. Survival rates generally decrease with lower passenger class, indicating a correlation between class and survival.

### Q 7: Does family size impact survival rates, considering gender?

```python
# create a plot family size vs survial
# Set the figure size
plt.figure(figsize=(16, 6))
# Create the bar plot with color
sns.barplot(x='family_size', y='survived', hue='who', data=df, palette='tab10')
# Set the title and labels
plt.title('Family Size vs. Survival')
plt.xlabel('Family Size')
plt.ylabel('Survival')
plt.show()
```

# Multivariate Analysis

### Q 8: What is the interaction between passenger class, age, and survival?

```python
# Class and Age vs. Survival
plt.figure(figsize=(16,6))
sns.catplot(x='pclass', y='age', hue='survived', kind='violin', data=df, height=6, aspect=1.5)
plt.title('Class and Age Interaction with Survival')
plt.show()
```

# Q 9: How do passenger class and family size together influence survival rates?

```python
# Class and Family Size vs. Survival
plt.figure(figsize=(16,6))
sns.heatmap(df.pivot_table('survived', 'family_size', 'pclass'), annot=True, cmap='YlGnBu')
plt.title('Class and Family Size Interaction with Survival')
plt.show()
```
